Criando o SparkSession

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
```

**CSV**

parametros importantes da leitura csv:

path: diretório onde localiza-se o arquivo.

Sep: Separador utilizado para separar as colunas.

Header: Booleano que considera a primeira linha como nome de colunas.

InferSchema: Infere o datatype baseado nas colunas disponiveis dentro dele


```python
df = spark.read.csv(path='/mnt/dados/clientes.csv',sep=';',header='True',inferSchema='True')
```

```python
df_endereco = spark.read.csv(path='/mnt/dados/endereco.txt',sep='|',header='True')
df_endereco.show()
```
