Lendo o arquivo desestruturado com Python

```python


linhas = []

nome_arquivos = ['06-09-2022.txt','05-09-2022.txt','04-09-2022.txt']

for nome_arquivo in nome_arquivos:
  arquivo = open('/mnt/{}'.format(nome_arquivo),'r')
  for linha in arquivo:
    linha_lista = linha.split('\t')
    if len(linha_lista) == 6:
      linhas.append(linha_lista)

print(linhas)

```
